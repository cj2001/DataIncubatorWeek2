```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
im2 = cv2.imread('image_b2.tif',0)
im5 = cv2.imread('image_b5.tif',0)
```

## LANDSAT Band 2 (Green) for region around Powder Mountain, Utah

```python
plt.imshow(im2, cmap = 'gray')
plt.colorbar()
```

##LANDSAT Band 5 (SWIR) for region around Powder Mountain, Utah

```python
plt.imshow(im5, cmap = 'gray')
plt.colorbar()
```

## Normalized Difference Snow Index (NDSI) for LANDSAT Imagery

$$NDSI = \frac{Band_{2} - Band_{5}}{Band_{2} + Band_{5}}$$

```python
base = np.ones(im2.shape)
im2n = im2 + base
im5n = im5 + base

num = np.absolute(im2n - im5n)
denom = im2n + im5n
ndsi = np.divide(num,denom)

plt.imshow(ndsi, cmap='gray')
plt.colorbar()
```

Studies have shown you can differentiate between snow and non-snow by thresholding NDSI > 0.4.  However, this can vary based on water content in the snow, grain size of snow, etc.  

```python
tndsi = ndsi > 0.4
plt.imshow(tndsi, cmap='gray')
plt.colorbar()
```

```python
tndsi = ndsi > 0.6
plt.imshow(tndsi, cmap='gray')
plt.colorbar()
```
